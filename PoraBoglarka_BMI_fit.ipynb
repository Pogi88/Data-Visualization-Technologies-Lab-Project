### **Könyvtárak és adatok importálása**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url_kaggle = 'https://raw.githubusercontent.com/vlewir/adatvizualizacio/refs/heads/main/BMI_fit_kaggle.csv'
url_own = 'https://raw.githubusercontent.com/vlewir/adatvizualizacio/refs/heads/main/BMI_fit_own.csv'
kaggle_data = pd.read_csv(url_kaggle)
own_data = pd.read_csv(url_own)
```

```python
kaggle_data.head()
```

```python
own_data.head()
```

### **Színskála**

```python
sns.set_style("white")
sns.set_palette("rocket")
```

```python
sns.color_palette()
```

```python
palette = sns.color_palette()
```

```python
label_color = '#474747'
```

### **Kaggle Adatok**

**Magasságok eloszlása**

```python
plt.figure(figsize=(12, 6))

ax = sns.histplot(kaggle_data['Magassag (m)'], bins=30, kde=False, color=palette[0])

plt.title('Magasságok Eloszlása', color=label_color, fontsize=12)
plt.xlabel('Magasság (m)', color=label_color)
plt.ylabel('')

plt.gca().spines['bottom'].set_color(label_color)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

ax.tick_params(axis='x', colors=label_color)
ax.tick_params(axis='y', colors=label_color)

plt.show()
```

**Testsúlyok eloszlása**

```python
plt.figure(figsize=(12, 6))
sns.histplot(kaggle_data['Suly (kg)'], bins=30, kde=False, color=palette[1])

plt.title('Testsúlyok Eloszlása', color=label_color, fontsize=12)
plt.xlabel('Testsúly (kg)', color=label_color)
plt.ylabel('')

plt.gca().spines['bottom'].set_color(label_color)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

ax.tick_params(axis='x', colors=label_color)
ax.tick_params(axis='y', colors=label_color)

plt.show()
```

**BMI eloszlása**




```python
plt.figure(figsize=(12, 6))
sns.histplot(kaggle_data['BMI'], bins=30, kde=False, color=palette[2])

plt.title('BMI Eloszlása', color=label_color, fontsize=12)
plt.xlabel('BMI', color=label_color)
plt.ylabel('')

plt.gca().spines['bottom'].set_color(label_color)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

ax.tick_params(axis='x', colors=label_color)
ax.tick_params(axis='y', colors=label_color)

plt.show()
```

**Magasság vs. Testsúly**

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=kaggle_data['Magassag (m)'], y=kaggle_data['Suly (kg)'], color=palette[3])

plt.title('Magasság vs Testsúly', color=label_color, fontsize=12)
plt.xlabel('Magasság (m)', color=label_color)
plt.ylabel('Testsúly (kg)', color=label_color)

plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.gca().tick_params(axis='x', colors=label_color)
plt.gca().tick_params(axis='y', colors=label_color)

plt.show()
```

**Magasság vs. BMI**

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=kaggle_data['Magassag (m)'], y=kaggle_data['BMI'], color=palette[4])

plt.title('Magasság vs BMI', color=label_color, fontsize=12)
plt.xlabel('Magasság (m)', color=label_color)
plt.ylabel('BMI', color=label_color)

plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

ax.tick_params(axis='x', colors=label_color)
ax.tick_params(axis='y', colors=label_color)

plt.show()
```

**Testsúly vs. BMI**

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=kaggle_data['Suly (kg)'], y=kaggle_data['BMI'], color=palette[5])

plt.title('Testsúly vs BMI', color=label_color, fontsize=12)
plt.xlabel('Testsúly (kg)', color=label_color)
plt.ylabel('BMI', color=label_color)

plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

ax.tick_params(axis='x', colors=label_color)
ax.tick_params(axis='y', colors=label_color)

plt.show()
```
